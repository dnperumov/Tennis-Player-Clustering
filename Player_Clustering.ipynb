```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing data
import os
directory = 'tennis_data2010'


if not os.path.exists(directory):
    os.makedirs(directory)

years = range(2010, 2025)
dfs = []
# save the file
for year in years:
    file_path = os.path.join(directory, f'atp_matches_{year}.csv')


    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
    else:
        url = f'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_{year}.csv'
        df = pd.read_csv(url)
        df.to_csv(file_path, index=False)

    dfs.append(df)

# single dataframe
```

```python
# winner and loser dataframes

w_cols = ['match_num', 'winner_id', 'winner_name','winner_hand', 'w_ace',
'w_df',
'w_svpt',
'w_1stIn',
'w_1stWon',
'w_2ndWon',
'w_SvGms',
'w_bpSaved',
'w_bpFaced',
'winner_ht',
'minutes',
'winner_rank'
]
l_cols = ['match_num', 'loser_id', 'loser_name', 'loser_hand', 'l_ace',
'l_df',
'l_svpt',
'l_1stIn',
'l_1stWon',
'l_2ndWon',
'l_SvGms',
'l_bpSaved',
'l_bpFaced',
'loser_ht',
'loser_rank',
'minutes']

# dataframes of the winners and losers of every match

winner_df = df[w_cols]
loser_df = df[l_cols]

winner_df.rename(columns = {'winner_id':'id'}, inplace = True)
loser_df.rename(columns = {'loser_id':'id'}, inplace = True)

winner_df.rename(columns = {'winner_name':'name'}, inplace = True)
loser_df.rename(columns = {'loser_name':'name'}, inplace = True)


all_players = pd.concat([winner_df, loser_df], ignore_index=True)


win_counts = winner_df['id'].value_counts()
loss_counts = loser_df['id'].value_counts()


players_with_5_wins = win_counts[win_counts >= 5].index
players_with_5_losses = loss_counts[loss_counts >= 5].index


players_with_5_wins_losses = set(players_with_5_wins) & set(players_with_5_losses)


f_winner_df = winner_df[winner_df['id'].isin(players_with_5_wins_losses)]
f_loser_df = loser_df[loser_df['id'].isin(players_with_5_wins_losses)]

# total matches
total_matches = win_counts.add(loss_counts, fill_value=0)

# win percentage
players_with_5_wins_losses = list(players_with_5_wins_losses)

# win percentage
win_percentage = win_counts.loc[players_with_5_wins_losses] / total_matches.loc[players_with_5_wins_losses]
all_players.head()
```

```python
# Combine the maximum (career-high) rank for winners and losers
max_winner_rank = f_winner_df.groupby('id')['winner_rank'].min()
max_loser_rank = f_loser_df.groupby('id')['loser_rank'].min()

max_rank = max_winner_rank.combine_first(max_loser_rank)
max_rank
#Merge the career-high rank information with the average stats data

```

```python
# Compute average stats without the 'name' column
w_average_stats = w_grouped[['winner_rank', 'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms', 'w_bpSaved', 'w_bpFaced', 'winner_ht', 'minutes']].mean()
l_grouped = f_loser_df.groupby(['id'])

l_average_stats = l_grouped[['loser_rank', 'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced', 'loser_ht', 'minutes']].mean()

# Merge back the 'name' column separately if needed
w_average_stats['name'] = w_grouped['name'].first()
l_average_stats['name'] = l_grouped['name'].first()
```

```python


#computing averages for each player
merged_stats = pd.merge(w_average_stats, l_average_stats, on = ['id'])

merged_stats['avg_ace'] = (merged_stats['w_ace'] + merged_stats['l_ace']) / 2
merged_stats['avg_df'] = (merged_stats['w_df'] + merged_stats['l_df']) / 2
merged_stats['avg_svpt'] = (merged_stats['w_svpt'] + merged_stats['l_svpt']) / 2
merged_stats['avg_1stIn'] = (merged_stats['w_1stIn'] + merged_stats['l_1stIn']) / 2
merged_stats['avg_1stWon'] = (merged_stats['w_1stWon'] + merged_stats['l_1stWon']) / 2
merged_stats['avg_2ndWon'] = (merged_stats['w_2ndWon'] + merged_stats['l_2ndWon']) / 2
merged_stats['avg_SvGms'] = (merged_stats['w_SvGms'] + merged_stats['l_SvGms']) / 2
merged_stats['avg_bpSaved'] = (merged_stats['w_bpSaved'] + merged_stats['l_bpSaved']) / 2
merged_stats['avg_bpFaced'] = (merged_stats['w_bpFaced'] + merged_stats['l_bpFaced']) / 2
merged_stats['height'] = (merged_stats['winner_ht'] + merged_stats['loser_ht']) / 2
merged_stats['minutes'] = (merged_stats['minutes_x'] + merged_stats['minutes_y']) / 2
merged_stats['first_serve_p'] = 100 * (merged_stats['avg_1stIn'] / merged_stats['avg_svpt'])
merged_stats['first_serve_won_p'] = 100 * (merged_stats['avg_1stWon'] / merged_stats['avg_1stIn'])
merged_stats['second_serve_won_p'] = 100 * (merged_stats['avg_2ndWon'] / (merged_stats['avg_svpt'] - merged_stats['avg_1stIn']))
merged_stats['ace_rate'] = merged_stats['avg_ace'] / merged_stats['avg_svpt']
merged_stats['bpsaved_perSvGm'] = merged_stats['avg_bpSaved'] / merged_stats['avg_SvGms']
merged_stats['bpfaced_perSvGm'] = merged_stats['avg_bpFaced'] / merged_stats['avg_SvGms']
merged_stats['df_rate'] = merged_stats['avg_df'] / merged_stats['avg_svpt']
merged_stats['pointsper_svgame'] = merged_stats['avg_svpt'] / merged_stats['avg_SvGms']



stats_cols = ['id', 'first_serve_p', 'first_serve_won_p', 'second_serve_won_p', 'avg_ace', 'avg_df', 'avg_bpSaved', 'avg_bpFaced','pointsper_svgame', 'win_percentage', 'height']

# add win percentage to stats df
win_percentage = win_percentage.reset_index()

win_percentage.columns = ['id', 'win_percentage']

merged_stats = pd.merge(merged_stats, win_percentage, on='id', how='left')

stats_only = merged_stats[['first_serve_p', 'first_serve_won_p', 'second_serve_won_p', 'avg_ace', 'avg_df', 'avg_bpSaved', 'avg_bpFaced','pointsper_svgame']]
stats_data = merged_stats[stats_cols]




```

```python
# k means clustering
from sklearn.cluster import KMeans
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#convert to array
arr = stats_only.values




inertia = []

# inertia for different values of k
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(arr)
    inertia.append(kmeans.inertia_)

# plot elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()


```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer


arr = stats_only.values

fig, ax = plt.subplots(3, 2, figsize=(15, 18))
for i in [2, 3, 4, 5]:
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(i, 2)

    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(arr)
```

```python
# running k means
# Using sklearn


cluster_labels = km.predict(arr)

stats_only['Cluster'] = cluster_labels

km = KMeans(n_clusters=5, random_state=3425)
km.fit(arr)

# statistical summary for each cluster
for cluster_id in range(5):
    cluster_data = stats_only[cluster_labels == cluster_id]
    print(f"Cluster {cluster_id} has {len(cluster_data)} data points.")

    print(cluster_data.describe())
    cluster_data_mean = cluster_data.mean()
```
